```python
#Basic Line plot
x= [1,2,3,4]
y=[11,24,36,45]
plt.plot(x,y);
```

```python
#1st Method
fig = plt.figure()#Creates a figure
ax= fig.add_subplot() #adds some axes
plt.show()
```

```python
#2nd Method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd method (recommended)
fig , ax = plt.subplots()
ax.plot([50,90,11,14,52]); #add some data
```

 ### Matplotlib Example workflow

```python
# 0. Import matplotlib and get it ready for plotting in juyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plots
fig , ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x,y)
                    
#4. Customise plot
ax.set(title = 'simple plot', 
      xlabel = ' x axis',
      ylabel = 'y axis');

# 5. Save & show (you save the whole plot)
#fig.savefig("Matplotlib/sample-plots.png")
```
