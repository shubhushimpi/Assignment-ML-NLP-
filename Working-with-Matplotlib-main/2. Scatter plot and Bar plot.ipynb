It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK. 

### Importing libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Making Figures with Numpy arrays

1. Line Plot
2. Scatter plots 
3. Bar Plot
4. Histogram
5. Subplots


```python
#Create some data
x = np.linspace(0,10,50)
x
```

```python
#Plot the data
fig , ax = plt.subplots()
ax.plot(x, x**2)
```

```python
#Scatter plot
fig , ax = plt.subplots()
ax.scatter(x,np.square(x));
```

```python
#Another Scatter plot
fig , ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
book_prices = {"Seven angles": 5,
              "Art of something": 8,
              "Pretty little liars":10}
fig, ax = plt.subplots()
ax.bar(book_prices.keys(),book_prices.values())
ax.set(title= 'Books And Prices',
      ylabel= ' prices');
```

```python
#Horizontal Bargraph
fig , ax = plt.subplots()
ax.barh(list(book_prices.keys()),list(book_prices.values()));
```

```python
#Histogram
x = np.random.randn(100)
fig, ax = plt.subplots()
ax.hist(x);
```

### Subplots (Two options)


```python
#SUBPLOTS OPTION 1
fig , ((ax0,ax1),(ax2,ax3)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
#Plot to each different axis
ax0.plot(x,x*2)
ax1.scatter(x,np.sin(x))
ax2.hist(x)
ax3.bar(book_prices.keys(),book_prices.values());
```
