### Manipulating Data with pandas

```python
car_sales["Make"].str.lower()  #Will change the make column to loweracase here but not in dataset
```

```python
car_sales
```

```python
car_sales["Make"] = car_sales["Make"].str.lower() #To Change it in dataset
```

```python
car_sales
```

```python
#Lets Import the Dataset with missing Values
car_sales_missing = pd.read_csv("car-sales-missing-data.csv")
```

```python
car_sales_missing
```

```python
# Fill Missing Values
# Use attribute inplace = True to fill the Values without Reassigning the Variable
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

#### we can see the missing values of odometer are fill with the mean value

```python
car_sales_missing.dropna()
```

```python
car_sales_missing #We still have those in dataset
```

```python
car_sales_missing.dropna(inplace = True)
```

```python
car_sales_missing
```

we can see all the nan are gone

```python
# Column from Series
seats_column = pd.Series([5, 5, 5, 5, 5])

# New Column called Seats
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True) 
```

```python
car_sales
```

```python
# Column with List
fuel_eco = [7.5,9.2,6.0,8.6,7.7,8.2,5.2,6.2,2,9]
car_sales["Fuel per 100KM"] = fuel_eco
```

```python
car_sales
```

```python
#Column using existed columns
car_sales["Total Fuel Used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```

```python
car_sales
```

```python
# Create a Column from a Single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed Road Safety"] = True
car_sales.dtypes
```

```python
#To delete a column
car_sales.drop("Total Fuel Used", axis=1, inplace=True) #axis = 1 is column
```

```python
car_sales
```

```python
# To randomly sample data
# to Randomly Order to the index
car_sales.sample(frac=0.5)
```

```python
car_sales = car_sales.sample(frac=1)
car_sales  #this randomly assign data with entire dataset
```

```python
#this randomly assign data with 20% of dataset
car_sales.sample(frac=0.2)
```

```python
car_sales
```

```python
# To reset the Index into the Original Format
# Use Attribute drop to avoid adding an additional column of index with random order
car_sales.reset_index(drop=True, inplace=True)
```

```python
car_sales
```

```python
# Applying Functions to the Columns
# Converting the Odometer column from KM to Miles
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x/1.6) 
car_sales
```
