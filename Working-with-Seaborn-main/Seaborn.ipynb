An introduction to seaborn
Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.
Seaborn helps you explore and understand your data. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots. Its dataset-oriented, declarative API lets you focus on what the different elements of your plots mean, rather than on the details of how to draw them.

### Imporing Seaborn library

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

###  =>correlation with Heatmap<br>=>corelation we find interdependence between all the features.<br>=>it applies to quantitative data only(having either interger or float values)

```python
sns.heatmap(df.corr(), annot=True);
```

```python
df.corr()
```

### Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df, kind='hex')
```

```python
# we can observe that most of the tip is between 2-4USD 
#total bill is in between 10-20USD
```

### Pair Plot

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue='sex', palette='husl')
```

```python
sns.pairplot(df,hue='smoker', palette='cubehelix')
```

### => Bar plot in seaborn takes both x and y values<br>=>Countplot can plot with either x or y value both<br>=>Are not necessary

```python
sns.barplot(x='total_bill', y='sex', data=df, palette='husl')
```

```python
sns.barplot(y='total_bill', x='smoker', data=df,palette='dark')
```

```python
sns.barplot(x='smoker', y='tip',hue='sex', palette='magma',data=df,capsize=.2)
```

```python
sns.barplot(x='tip', y='sex', hue='time',palette='winter',data=df)
```

### =>we observe that most tips are received during dinner time both by males and females<br>=>we can also infer that people dine out more at nights.

```python
sns.barplot(x='day',y='tip',hue='smoker',data=df, palette='Blues')
```

### Box Plot

```python
sns.boxplot('smoker', 'total_bill',data=df,palette='husl')
```

```python
sns.boxplot('sex', 'total_bill',data=df,palette='Greens')          
```

```python
sns.boxplot('day','total_bill',data=df, palette='Accent')
```

```python
sns.boxplot('day','total_bill',data=df,hue='smoker',palette='Pastel2')
```

### =>we observe that maximum outliers are for non-smokers on saturday<br>=>sunday smokers we fine no outliers, almost symmetrical boxplot

### Violin Plots
Violin plots helps to see the data both in terms of kernel density estimation

```python
sns.violinplot(x='sex', y='total_bill',data=df,palette='rocket')
```

### =>Majority Males pay bill between 10-20USD. Bill received is maximum in case of males<br>=>Majority Females pay bill between 7-15USD, Bill received has not reached the maximum value

```python
sns.violinplot(x='total_bill',y='day',data=df,hue='sex', palette='terrain')
```

### =>we observe most outliers on saturday.<br>=>Clearly see how data is distributed or spread.<br>=>Most bill range is between 10-20USD

```python
sns.violinplot('tip','day', hue='smoker',data=df, palette='seismic')
```

```python
sns.violinplot('total_bill','day',data=df, hue='time',palette='Accent')
```

### =>Total bill received is maximum on saturay and sunday during dinner time.<br>=>And since there is a positive corelation between total bill and tip<br>=>Tip will also increase during dinner time in weekends.

```python
sns.violinplot('tip','day',palette='Wistia',data=df)
```

```python
sns.violinplot('tip','sex',palette='tab20_r',data=df,hue='time')
```

### Scatter Plots

```python
sns.catplot(x='day',y='tip', data=df)
```

```python
sns.catplot(x='day',y='tip', data=df, jitter=False) 
#notice jitter this reduces the magnitude of jitter in the parameter
```

```python
sns.catplot(x='sex',y='tip',kind='swarm',data=df)
```

We observe that most of the tips are paid between $2-$4 
males paying most $2 tip than females.

```python
sns.catplot(x='sex', y='total_bill', kind='swarm',data=df, palette='husl');
```

### =>We observe that most of the males pay the bill between $10-$20 <br>=>This also applies for females. Also a few males pay the highest bills compared to females.

```python
sns.catplot(x='sex',y='tip',hue='smoker', data=df, palette='winter',
            kind='swarm')         
```

```python
sns.catplot(x='smoker',y='tip',palette='rainbow',hue='sex',data=df, kind='swarm');
```

```python
sns.catplot('smoker','tip',order=['No','Yes'], data=df)
```

### =>We observe that there are more male smokers who pay the tip<br>=>For non-smokers also males pay the most tip <br>=>We can also infer that males both smoker and non smoker pay the maximum tips.

```python
sns.catplot('tip','sex',hue='time',palette='coolwarm',data=df)
```
